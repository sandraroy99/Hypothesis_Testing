Q1

Here we can perform two tailed t test.
 firstly we need to check the normality of the data and the homogenity of the variance.
 


```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats  as stat
```

```python
df=pd.read_csv(r"C:\Users\sandr\Downloads\Cutlets.csv")
df
```

```python
column_a=df['Unit A']
column_b=df['Unit B']
```

```python
#Normality check using shapiro
from scipy.stats import shapiro
stat1, p_value1 =shapiro(column_a)
stat2, p_value2 =shapiro(column_b)
print(f'statistic:{stat1},p-value:{p_value1}')
print(f'statistic:{stat2},p-value):{p_value2}')
if p_value1 and p_value2>0.05:
    print('The data appears to be normally distributed.')
else:
    print('The data does not appers to be normally distributed')
```

```python
#checking Homogenity of variance using Levene's test and bartlett test
from scipy.stats import levene
from scipy.stats import bartlett
statistic1,p_value1 =levene(column_a,column_b)
statistic2,p_value2 =bartlett(column_a,column_b)
print(f'levenes statistic:{statistic1},p-value:{p_value1}')
print(f'bartlett statistic:{statistic2},p-value:{p_value2}')
if p_value1 and p_value2  <0.05:
    print('variance are  not equal')
else:
    print('variance are equal ')

```

```python
#Performing t test
#H0:There is no significant difference in the mean diameter between two units.
#H1: There is a significant difference in the mean diameter between two units.
from scipy.stats import ttest_ind
statistic, p_value= ttest_ind(column_a,column_b)
print(f'p-value:{p_value}')
if p_value <0.05:
    print('we are rejecting the null hypothesis')
else:
    print('we are accepting the null hypothesis')

```

```python
df1=pd.read_csv(r"C:\Users\sandr\Downloads\LabTAT.csv")
df1
```

```python
lab1=df1['Laboratory 1']
lab2=df1['Laboratory 2']
lab3=df1['Laboratory 3']
lab4=df1['Laboratory 4']
```

```python

#Normality check using shapiro
from scipy.stats import shapiro
stat1, p_value1 =shapiro(lab1)
stat2, p_value2 =shapiro(lab2)
stat3, p_value3 = shapiro(lab3)
stat4, p_value4 =shapiro(lab4)
print(f'statistic:{stat1},p-value:{p_value1}')
print(f'statistic:{stat2},p-value):{p_value2}')
print(f'statistic:{stat3},p-value):{p_value3}')
print(f'statistic:{stat4},p-value):{p_value4}')
if p_value1 and p_value2 and p_value3 and p_value4>0.05:
    print('The data appears to be normally distributed.')
else:
    print('The data does not appers to be normally distributed')
```

```python
#checking Homogenity of variance using Levene's test and bartlett test
from scipy.stats import levene
from scipy.stats import bartlett
statistic1,p_value1 =levene(lab1,lab2,lab3,lab4)
statistic2,p_value2 =bartlett(lab1,lab2,lab3,lab4)
print(f'levenes statistic:{statistic1},p-value:{p_value1}')
print(f'bartlett statistic:{statistic2},p-value:{p_value2}')
if p_value1 and p_value2  <0.05:
    print('variance are  not equal')
else:
    print('variance are equal ')

```

```python
#Performing one way annova
#H0:There is no significant difference in the mean TAT among laboratories.
#H1: There is a significant difference in the mean TAT among laboratories.
from scipy.stats import f_oneway

statistic, p_value= f_oneway(lab1,lab2,lab3,lab4)
print(f'One way annova statistic:{statistic},p-value:{p_value}')
if p_value <0.05:
    print('we are rejecting the null hypothesis')
else:
    print('we are accepting the null hypothesis')
```
